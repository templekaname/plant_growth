```python
%load_ext autoreload
%autoreload 2

import numpy as np
from growth_model import *

pi = np.pi
```

```python
# Parameters
E = 48*1e3*1e3 # youngs modulus (gf/cm^2)
rho = 0.82 # density (o.82 gf/cm^3)
r_0 = 0.134 # initial radius (cm)
dr = 0.0213 # radius increment (cm)
dl = 5 # length increment (cm)
dh = 0 # foliage weight increment at internodes (gf)
h_end = 0 # foliage weight at tip (gf)
theta_0 = pi/6 # initial angle
theta_C = "straight" # preferred angle at end
theta_P = pi/2 # preferred angle at lignified stem
rw_stress = 0 # recovery moment (gf/cm^2)
N = 80 # Number of the growth step
```

```python
pm = ParamPlant(E, rho, r_0, dr, dl, rw_stress)
pf = ParamFoliage(dh, h_end)
pa = ParamAngle(theta_0, theta_C, theta_P)
model = GrowthModel(N, pm, pf, pa)
```

```python
model.simulate()
```

```python
model.track_growth(np.arange(10,81,10))
```
